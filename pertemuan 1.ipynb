Apa itu ERD?

ERD (Entity Relationship Diagram) adalah diagram yang digunakan untuk menggambarkan struktur data dalam sebuah sistem.
ERD menunjukkan:

Entitas (objek)

Atribut (informasi)

Relasi antar entitas

Kardinalitas (jenis hubungan)


ERD sangat penting sebagai langkah desain sebelum membuat tabel pada basis data.


---

ğŸ“Œ Komponen Utama ERD

1. Entitas

Objek nyata yang datanya ingin disimpan.
Contoh pada sistem apotek:

Pasien

Pegawai

Obat

Resep

Pembayaran



---

2. Atribut

Informasi yang dimiliki entitas.
Contoh atribut Pasien:

Nama

Alamat

Tanggal Lahir

No RM (sebagai PK)



---

3. Primary Key (PK)

Atribut yang unik untuk membedakan setiap baris data.
Contoh: KodeObat, IDPegawai, NoResep.


---

4. Relasi

Hubungan antar entitas.
Contoh:

Pasien memiliki Resep

Pegawai melayani Pembayaran

Resep berisi Obat



---

5. Kardinalitas

Menjelaskan jumlah hubungan:

1 : 1 â†’ Satu dengan satu

1 : N â†’ Satu dengan banyak

N : M â†’ Banyak dengan banyak

4. Weak Entity â†’ PK Gabungan

5. Relasi 1 : 1 â†’ Tambah Foreign Key

Salah satu tabel menyimpan FK dari tabel lainnya.

6. Relasi 1 : N â†’ FK di sisi N

Tabel yang berada di posisi N menerima FK.
